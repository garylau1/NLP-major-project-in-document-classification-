```python
import pandas as pd

# 1. Read in your CSV
df = pd.read_csv('latest_metrics_summary_claude.csv')

# 2. Group by the class and compute mean F1 for PII, non-PII, and combined
grouped_claude = (
    df
    .groupby('top_level_key')
    .agg(
        avg_f1_pii      = ('f1_pii',      'mean'),
        avg_f1_non      = ('f1_non',      'mean'),
        avg_f1_combined = ('f1_combined', 'mean')
    )
    .reset_index()
)

# 3. Display the result
grouped_claude
```

```python
import pandas as pd

# 1. Read in your CSV
df = pd.read_csv('latest_metrics_summary_llema_4.csv')

# 2. Group by the class and compute mean F1 for PII, non-PII, and combined
grouped_llema_4 = (
    df
    .groupby('top_level_key')
    .agg(
        avg_f1_pii      = ('f1_pii',      'mean'),
        avg_f1_non      = ('f1_non',      'mean'),
        avg_f1_combined = ('f1_combined', 'mean')
    )
    .reset_index()
)

# 3. Display the result
grouped_llema_4
```

```python
import pandas as pd

df_piiranha = pd.read_csv("latest_metrics_summary_piiranha.csv")


df_piiranha["avg_f1_combined"]=df_piiranha["f1"]

df_piiranha["top_level_key"]=df_piiranha["class_name"]

```

```python
df_piiranha
```

```python
import matplotlib.pyplot as plt

# 1. Pull out and rename just the combined-F1 column from each summary
claude = grouped_claude[['top_level_key', 'avg_f1_combined']].rename(columns={'avg_f1_combined':'Claude'})
llema  = grouped_llema_4[['top_level_key', 'avg_f1_combined']].rename(columns={'avg_f1_combined':'Llama'})
piira  = df_piiranha[['top_level_key', 'avg_f1_combined']].rename(columns={'avg_f1_combined':'Piiranha'})

# 2. Merge them all on the class key
df_compare = (
    claude
    .merge(llema,  on='top_level_key')
    .merge(piira,  on='top_level_key')
    .set_index('top_level_key')
)

```

```python
import matplotlib.pyplot as plt

# 1. Create a larger figure
fig, ax = plt.subplots(figsize=(20, 8))   # make the canvas bigger

# 2. Plot on that axis
df_compare.plot(kind='bar', ax=ax)

# 3. Tweak the text sizes
ax.set_title('F1 Comparison: Claude vs Llama vs Piiranha', fontsize=16)
ax.set_ylabel('Average Combined F1 Score', fontsize=14)

# 4. Rotate & shrink x-tick labels
ax.set_xticklabels(
    df_compare.index,
    rotation=45,
    ha='right',
    fontsize=8  # smaller x-tick font
)

# 5. Shrink y-tick labels and legend
ax.tick_params(axis='y', labelsize=10)
ax.legend(fontsize=10)

plt.tight_layout()
plt.show()
```
